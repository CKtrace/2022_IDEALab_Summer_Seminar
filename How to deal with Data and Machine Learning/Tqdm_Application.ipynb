```python
import numpy as np
import matplotlib.pyplot as plt
import tqdm
import time

bream_length = np.array([25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0])
bream_weight = np.array([242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0])

# y = ax + b
a = 1
b = 0

lr = 0.0005

epochs = 1000000
for i in range(100):    
    y_pred = a * bream_length + b
    error = bream_weight - y_pred
    
    a_diff = -(2/len(bream_length)) * sum((bream_length*(error)))
    b_diff = -(2/len(bream_weight)) * sum(error)
    
    a = a - lr * a_diff
    b = b - lr * b_diff
    for z in tqdm.tqdm(range(epochs//100), desc = "Process", mininterval=0.01):
        y_pred = a * bream_length + b
        error = bream_weight - y_pred
        
        #MSE
        a_diff = -(2/len(bream_length)) * sum((bream_length*(error)))
        b_diff = -(2/len(bream_weight)) * sum(error)
        
        a = a - lr * a_diff
        b = b - lr * b_diff
#         time.sleep(0.00001)
    
    print("epoch = %.f, 기울기 = %.04f,  절편 = %.04f, 오차=%.04f" % ((i*epochs//100)+10000, a, b, error.mean()))
    print(a_diff)
    print(b_diff)    
        

plt.title("MSE with TQDM")
plt.scatter(bream_length, bream_weight)
plt.plot([min(bream_length), max(bream_length)], [min(y_pred), max(y_pred)], 'g')
plt.xlabel('Length')
plt.ylabel('Weight')
plt.show()
```
