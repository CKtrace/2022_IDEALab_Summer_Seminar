```python
import matplotlib.pyplot as plt
import numpy as np


def vector (a, b) :
    
    vec = [b[0]-a[0], b[1]-a[1], b[2]-a[2]]
    
    return vec


def outer_product (a, b) :
    
    x = a[1]*b[2] - a[2]*b[1]
    y = -(a[0]*b[2] - a[2]*b[0])
    z = a[0]*b[1] - a[1]*b[0]
    q = -(a[0]*x + a[1]*y + a[2]*z)
    
    new_arr = [x,y,z,q]
    
    return new_arr


fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(1, 1, 1, projection='3d')


ax.set_xlabel("x", size = 12) 
ax.set_ylabel("y", size = 12) 
ax.set_zlabel("z", size = 12)

ax.set_xlim([-20.0, 20.0])
ax.set_ylim([-20.0, 20.0])
ax.set_zlim([-20.0, 20.0])


ax.set_xticks([-20.0, -10.0, 0.0, 10.0, 20.0])
ax.set_yticks([-20.0, -10.0, 0.0, 10.0, 20.0])
ax.set_zticks([-20.0, -10.0, 0.0, 10.0, 20.0])


p = (2, -1, 1)
q = (4, -2, 4)
r = (5, -3, 2)


vecA = vector(p, q)
vecB = vector(p, r)


ans = outer_product(vecA, vecB)


xx, yy = np.meshgrid(range(-10, 10), range(-10, 10))


zz = (-ans[0]*xx - ans[1]*yy - ans[3]) / ans[2]


surf = ax.plot_surface(xx,yy,zz, alpha=0.3)


ax.plot(p[0],p[1],p[2], linestyle="none", marker="o", mfc="none", markeredgecolor="red")
ax.plot(q[0],q[1],q[2], linestyle="none", marker="o", mfc="none", markeredgecolor="red")
ax.plot(r[0],r[1],r[2], linestyle="none", marker="o", mfc="none", markeredgecolor="red")


ax.quiver(p[0], p[1], p[2], ans[0], ans[1], ans[2])


plt.show()
```
