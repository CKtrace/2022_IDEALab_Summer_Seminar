```python
import matplotlib.pyplot as plt
import numpy as np

plt.hist(np.random.randint(1, 4, 100000), bins = 15, color = 'g', alpha = 0.5, edgecolor = 'black')
#np.random.randint(align = 'mid')는 디폴트
plt.show()
```

```python
def mean_func(a, b) :
    """
        
    """
    vector_matrix = np.random.randint(1, 3, size = (5,2))
#     print(vector_matrix[a])
#     print(vector_matrix[b])
    
#     print(dot_product(vector_matrix[a], vector_matrix[b]))
    return dot_product(vector_matrix[a], vector_matrix[b])
    
def dot_product(v1, v2) :
    """
        v_1 * w_1 + ... + v_n * w_n  
    """
    
    return sum(v1_i * v2_i for v1_i, v2_i in zip(v1, v2))
    
a = np.array([])

for i in range(100000):
    rand_num1 = np.random.randint(0,5)
    rand_num2 = np.random.randint(0,5)
    value = mean_func(rand_num1, rand_num2)
    a = np.append(a, value)
    
plt.hist(a[:], bins = 15, color = 'g', alpha = 0.5, edgecolor = 'black')
plt.show()
```
